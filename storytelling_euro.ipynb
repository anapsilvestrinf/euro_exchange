# Storytelling Data Visualization on Exchange Rates

## Introduction

In this project, we will work with daily data that describes the Euro exchange rate between 1999 and 2021. This dataset is available for download [here](https://www.kaggle.com/lsind18/euro-exchange-daily-rates-19992020).

Our goal is to create a storytelling data visualization on the exchange rate between the euro and the American dollar. 

## Reading and Exploring data

First, we will understand the structure of the dataset.

```python
# Read file
import pandas as pd
exchange_rates = pd.read_csv('euro-daily-hist_1999_2020.csv')
# First five rows
display(exchange_rates.head(5))
# Last five rows
display(exchange_rates.tail(5))

# Facts about the dataset
display(exchange_rates.info())
```

## Data Cleaning

We will clean our data to make our work easier later. Our data cleaning involves:
* Renaming the \[US dollar \] and Period\Unit columns to something easier to type.
* Changing the Period\Unit column to a DateTime data type.
* Sorting the values by Period\Unit in ascending order.
* Resetting the index.
* Isolating columns.
* Dropping some rows.
* Converting \[US dollar \] to float.

```python
# Rename [US dollar ] and Period\Unit columns 
exchange_rates.rename(columns={'[US dollar ]': 'US_dollar',
                               'Period\\Unit:': 'Time'},
                      inplace=True)
# Convert Period\Unit to datetime
exchange_rates['Time'] = pd.to_datetime(exchange_rates['Time'])

#Sort values by Time
exchange_rates.sort_values('Time', inplace=True)

# Reset index
exchange_rates.reset_index(drop=True, inplace=True)

# Isolate Time and US_dollar
euro_to_dollar = exchange_rates[['Time', 'US_dollar']].copy()
display(euro_to_dollar['US_dollar'].value_counts())

# Drop rows with - character in US_dollar column
euro_to_dollar =  euro_to_dollar[euro_to_dollar['US_dollar'] != '-']

# Convert US_dollar to float
euro_to_dollar['US_dollar'] = euro_to_dollar['US_dollar'].astype(float)
```

## Rolling Mean

We will create a line plot to visualize the evolution of the euro-dollar exchange rate.

```python
# Plot chart
import matplotlib.pyplot as plt
%matplotlib inline 

rolling_mean = euro_to_dollar['US_dollar'].rolling(30).mean()

plt.plot(euro_to_dollar['Time'],
         rolling_mean)
plt.show()
```
